```python
import torch
import matplotlib.pyplot as plt
import numpy as np
from pytorch_grad_cam import GradCAM
from pytorch_grad_cam.utils.image import show_cam_on_image

# Wrapper model
class CSRNetCountWrapper(torch.nn.Module):
    def __init__(self, model):
        super().__init__()
        self.model = model

    def forward(self, x):
        density_map = self.model(x)
        count = density_map.sum(dim=(1, 2, 3))
        return count.unsqueeze(-1)

# take 1 sample from data loader
sample_batch, target_batch = next(iter(train_loader))

input_tensor = sample_batch[0].unsqueeze(0)        # [1, C, H, W]
density_gt    = target_batch[0].cpu().numpy()      # [H, W]

# target layer
target_layer = model.frontend[-1]

# device selection
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
input_tensor = input_tensor.to(device)
wrapped_model = CSRNetCountWrapper(model).to(device).eval()

# initialize grad cam
cam = GradCAM(model=wrapped_model, target_layers=[target_layer])

# compute CAM
grayscale_cam = cam(input_tensor=input_tensor)[0]

# Preparing original image
img = input_tensor[0].cpu().numpy().transpose(1, 2, 0)

# Take only 0 - 2 channel since model has 4 channel
if img.shape[2] > 3:
    img = img[:, :, :3]

img = (img - img.min()) / (img.max() - img.min() + 1e-8)

# Overlay CAM over images
cam_image = show_cam_on_image(img.astype(np.float32), grayscale_cam, use_rgb=True)

plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.title("Original Image")
plt.imshow(img)
plt.axis("off")

plt.subplot(1, 2, 2)
plt.title("Grad-CAM")
plt.imshow(cam_image)
plt.axis("off")
plt.show()
```
